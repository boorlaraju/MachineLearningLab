```python
sns.histplot(df["SepalLengthCm"], bins=20, color="orange",kde=True)
```

```python
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"], color="green") 
plt.xlabel("SepalLength")
plt.ylabel("SepalWidth")
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x=df["SepalLengthCm"], y=df["SepalWidthCm"], color="green")
```

```python
sns.set(style="whitegrid")
sns.scatterplot(df, color="green")
```

```python
g = sns.FacetGrid(df, hue="Species")
g.map_dataframe(sns.scatterplot, x='SepalLengthCm', y='SepalWidthCm')
g.add_legend()
plt.show()
```

```python
species = df["Species"].unique()
list(species)
```

```python
sns.boxplot(x=df["Species"], y=df["PetalWidthCm"])
```

```python
species = df['Species'].unique()
data = [df[df['Species'] == sp]['PetalWidthCm'] for sp in species]

# Create boxplot
plt.boxplot(data, labels=species)

# Titles and labels
plt.title("Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.grid(True)

plt.show()
```
