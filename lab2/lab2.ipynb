```python
df = df.drop_duplicates()
```

```python
#Drop the column with most missing values

df = df.drop(columns=df.isnull().sum().idxmax())
```

```python
#Drop the rows with categorical missing values

categorical_cols = df.select_dtypes(include='object').columns
df = df.dropna(subset=categorical_cols)
```

```python
#Impute the rows with numerical missing values

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
```

```python
#
#Sort data by price and find most & least expensive cars

df_sorted = df.sort_values(by='Price in thousands')
least_expensive = df_sorted.head(1)
most_expensive = df_sorted.tail(1)
```

```python
#Write a function to find min and max of any column

def find_min_max(col):
    return df[col].min(), df[col].max()

```

```python
#Call the function for horsepower, length, fuel efficiency

find_min_max('Horsepower')
find_min_max('Length')
find_min_max('Fuel efficiency')
```

```python
cols_to_convert = ['Price in thousands', 'Sales in thousands', 'Horsepower', 'Fuel efficiency', 'Length']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```
